# Archetype Analysis

```python
import archetypes as arch
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd
import plotly.express as px
```

```python
random_state = np.random.RandomState(0)

wine = load_iris()

X = wine.data
target = wine.target
```

```python
aa = arch.AA(n_archetypes=3, n_init=10, max_iter=10_000, tol=1e-16, random_state=random_state)
aa.fit(X)

aa.archetypes_
```

```python
df = pd.DataFrame(aa.alphas_, columns=["Arch0", "Arch1", "Arch2"])
df["Target"] = target
df = df.astype({"Target": str})

fig = px.scatter_ternary(df, a="Arch0", b="Arch1", c="Arch2",
                         color="Target",
                         color_discrete_map = {"Arch0": 0, "Arch1": 1, "Arch2": 2})
fig.show()
```
